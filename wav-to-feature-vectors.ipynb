# Loading all data test

```python
path = '/content/drive/MyDrive/UTEC/CURSOS/2024.1/MACHINE LEARNING/PRACTICA/DataSet'+ "/train/"

# List to store the feature vectors
feature_vectors = []

# Iterate through each file in the 'train' folder
for filename in os.listdir(path):
    if filename.endswith(".wav"):
        file_path = os.path.join(path, filename)

        y, sr = librosa.load(file_path)

        mfccs = librosa.feature.mfcc(y=y, sr=sr)

        feature_vector = np.array(mfccs.mean(axis=1))

        feature_vectors.append([filename] + feature_vector.tolist())

        print(f"Processed and saved feature vector for {filename}")

# Convert the list to a DataFrame
df = pd.DataFrame(feature_vectors, columns=["filename"] + [f"mfcc_{i+1}" for i in range(len(feature_vectors[0])-1)])

# Save the DataFrame to a CSV file
df.to_csv("feature_vectors.csv", index=False)

print("All feature vectors saved to feature_vectors.csv")
```

# MERGE FEATURE VECTOR - LABELS

```python

# Define the paths to your CSV files
feature_vectors_path = "/content/feature_vectors.csv"
train_csv_path = "/content/train.csv"

# Read the CSV files into DataFrames
feature_vectors_df = pd.read_csv(feature_vectors_path)
train_df = pd.read_csv(train_csv_path)

train_df['filename'] = train_df['idx'].astype(str) + '.wav'

merged_df = pd.merge(feature_vectors_df, train_df[['filename', 'class']], on='filename', how='inner')

merged_df.to_csv("merged_feature_vectors_test.csv", index=False)

print("Merged feature vectors saved to merged_feature_vectors.csv")
```
